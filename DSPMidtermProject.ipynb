```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Uploaded CSV File.
data = pd.read_csv('123.csv')

# Selecting the date column from the csv file, pd.to_datetime to convert and lastly to select the column of the date.
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

# Graph Creation
plt.figure(figsize=(25,10))
plt.plot(data['temperature'], label='temperature', color = 'red')
plt.plot(data['humidity'], label='humidity', color = 'black')
plt.title('Graph of Temperature and Humidity over Time')
plt.xlabel('Time')
plt.ylabel('Value of The Temperature and Humidity')
plt.grid()
plt.legend()
plt.show()
```
